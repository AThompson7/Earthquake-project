```python
earthquake_path = "Resources/earthquakes.csv"
earthquake_data = pd.read_csv(earthquake_path)
earthquake_data
```

```python
earthquake_data_dropped = earthquake_data.drop(earthquake_data.columns[[5,6,7,8,9,10,11,12,14,15,16,17,18,19,20,21]],axis=1)
```

```python
earthquake_data_dropped
```

```python
earthquake_data_renamed = earthquake_data_dropped.rename(columns={"time":"Time", "latitude":"Latitude", "longitude":"Longitude", "depth":"Depth(km)", "mag":"Magnitude", "place": "Location"})
```

```python
earthquake_data_renamed
```

```python
earthquake_round = earthquake_data_renamed.round(2)
earthquake_clean = earthquake_round.dropna()
earthquake_clean
```

```python
# get country list to for loop
locations=earthquake_clean['Location']

cols=earthquake_clean.columns.to_list()

#get headings list to for loop
headings=[]
for i in range(1,len(cols)):
    headings.append(cols[i])
headings

collector=[]

try:
    for i in range(0,len(locations)):
        for j in range(0,len(headings)):
            

            location = earthquake_clean.loc[(earthquake_clean['Location']==locations[i]),[headings[j]]].squeeze()
            
            magnitude = earthquake_clean.loc[(earthquake_clean['Magnitude']==locations[i]),[headings[j]]].squeeze()

            depth = earthquake_clean.loc[(earthquake_clean['Depth(km)']==locations[i]),[headings[j]]].squeeze()
            
            latitude = earthquake_clean.loc[(earthquake_clean['Latitude']==locations[i]),[headings[j]]].squeeze()
            
            longitude = earthquake_clean.loc[(earthquake_clean['Longitude']==locations[i]),[headings[j]]].squeeze()
            
            time = earthquake_clean.loc[(earthquake_clean['Time']==locations[i]),[headings[j]]].squeeze()
            
            collection = {'Index':i,'Place':locations[i],'Magnitude':str(magnitude),'Time':str(time),'Depth':str(depth),'Longitude':str(longitude),'Latitude':str(latitude)}
                         
            collector.append(collection)

except Exception as e:
        print(e)
            

```
